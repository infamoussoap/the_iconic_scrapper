```python
html = urlopen('https://www.theiconic.com.au/blake-linen-blend-blazer-1191155.html')
bsObj = BeautifulSoup(html)
```

```python
bsObj.findAll({"h1", "h2", "h3", "h4", "h5", "h6"})
```

```python
for i in range(1, 6 + 1):
    print(f'Header {i}')
    print(bsObj.findAll(f'h{i}'))
    print('----------------------')
```
